```python
#avg number of null values in each row that contains a null value in the column i
nan_dict={}
for i in df.columns:
    nan_dict[i] = df.loc[df[i].isnull()].isnull().sum(axis=1).mean()
print(nan_dict)
```

```python
df_reduce_nan = df
for key,val in nan_dict.items():
    if val > 2:
        df_reduce_nan.drop(key, axis=1, inplace=True)

```
